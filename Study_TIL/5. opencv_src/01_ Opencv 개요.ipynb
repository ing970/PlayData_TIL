## ë™ì˜ìƒ ì½ê¸°
- VideoCapture í´ëž˜ìŠ¤ì‚¬ìš©
    - ê°ì²´ ìƒì„±
        - VideoCapture('ë™ì˜ìƒíŒŒì¼ ê²½ë¡œ'): ë™ì˜ìƒíŒŒì¼  // <span style='background-color:#fff5b1'>**( ë™ì˜ìƒíŒŒì¼ ê²½ë¡œëŠ” ë¬¸ìžì—´ )**
        - VideoCapture(ì›¹ìº  ID): ì›¹ìº   // <span style='background-color:#fff5b1'>**( ì›¹ìº  IDëŠ” ì •ìˆ˜ì—´ )**
- VideoCaptureì˜ ì£¼ìš” ë©”ì†Œë“œ
    - isOpened(): bool
        - ìž…ë ¥ ëŒ€ìƒê³¼ ì—°ê²°ë˜ì—ˆëŠ”ì§€ ì—¬ë¶€ë°˜í™˜
    - <span style='background-color:#fff5b1'>read(): (bool, img)
        - Frame ì´ë¯¸ì§€ë¡œ ì½ê¸°
        - ë°˜í™˜ê°’
            - bool: ì½ì—ˆëŠ”ì§€ ì—¬ë¶€
            - img: ì½ì€ ì´ë¯¸ì§€(ndarray)

#### ì›¹ìº 

```python
import cv2 

cap = cv2.VideoCapture(0)           # ì •ìˆ˜: web ìº  ì—°ê²°. => 0ë¶€í„° 1ì”© ì¦ê°€í•˜ëŠ” ê°’ì„ ì‚¬ìš©. ì›¹ìº ì´ ì—¬ëŸ¬ê°œì¸ ê²½ìš° ì§€ì •.
print(cap.isOpened())
while True:
    # í•œê°œ Frameì„ ì½ê¸°
    success, image = cap.read()    # í•œ í”„ë ˆìž„ì„ ì½ëŠ”ë‹¤.
    if not success:
        print('í”„ë ˆìž„ì„ ì½ì–´ì˜¤ì§€ ëª»í•¨')
        break                       # ì¢…ë£Œ
    
    # ì½ì€ Frameì„ ì¶œë ¥
    image = cv2.flip(image, 1)      # ì–‘ìˆ˜: ìˆ˜í‰ë°˜ì „ // 0: ìˆ˜ì§ë°˜ì „ // ìŒìˆ˜: ìˆ˜í‰/ìˆ˜ì§ ëª¨ë‘ ë°˜ì „
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # BGR => GRAY SCALE
    cv2.imshow('frame', image)
    if cv2.waitKey(1) == 27:        # 1000ë¶„ì— 1ì´ˆë§ˆë‹¤ 1í”„ë ˆìž„ì„ ì°ì–´ë‚¸ë‹¤ // 27ì€ esc key ==> ì¢…ë£Œ
        break

        
cap.release()                       # ì—°ê²° ë‹«ê¸°(í•´ì œ)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import cv2
# VideoCapture ê°ì²´ ìƒì„± => ë™ì˜ìƒíŒŒì¼ ê²½ë¡œë¥¼ ë¬¸ìžì—´ë¡œ ì „ë‹¬.
cap = cv2.VideoCapture('images/wave.mp4') # ì •ìˆ˜: web cam ì—°ê²°, ë¬¸ìžì—´: ë™ì˜ìƒíŒŒì¼ ì—°ê²°
print('ì—°ê²°ì„±ê³µì—¬ë¶€:',cap.isOpened())

FPS = cap.get(cv2.CAP_PROP_FPS)
DELAY = int(1000/FPS) # waitKey()ëŠ” intê°’ì„ ëŒ€ê¸°ì‹œê°„ìœ¼ë¡œ ë°›ëŠ”ë‹¤.

while True:
    
    success, image = cap.read() # í•œ Frame ì½ê¸°
    if not success:
        print('ì½ê¸° ì‹¤íŒ¨')
        break # ì¢…ë£Œ
    
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # í‘ë°±ìœ¼ë¡œ ë³¼ ì‹œ image => gray scale
    
    cv2.imshow('frame', image)
    if cv2.waitKey(DELAY) == 27:
        break


cap.release()  # ì—°ê²° ë‹«ê¸°(ëŠê¸°)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

### ë™ì˜ìƒ ì €ìž¥
- capture(read)í•œ ì´ë¯¸ì§€ í”„ë ˆìž„ì„ ì—°ì†ì ìœ¼ë¡œ ì €ìž¥í•˜ë©´ ë™ì˜ìƒ ì €ìž¥ì´ ëœë‹¤.
- VideoWriter ê°ì²´ë¥¼ ì´ìš©í•´ ì €ìž¥
    - **`VideoWriter(filename, codec, fps, size)`**
        - filename: ì €ìž¥ê²½ë¡œ
        - codec
            - VideoWriter_fourcc ì´ìš©
            
        - fps
            - FPS (Frame Per Second) - ì´ˆë‹¹ ëª‡ í”„ë ˆìž„ì¸ì§€ ì§€ì •
        - size
            - ì €ìž¥í•  frame í¬ê¸°ë¡œ ì›ë³¸ ë™ì˜ìƒì´ë‚˜ ì›¹ìº ì˜ width, height ìˆœì„œë¡œ ë„£ëŠ”ë‹¤.
    - `VideoWriter().write(img)`
        - Frame ì €ìž¥

```python
def func(*args, **kwargs):
    print(args, type(args), len(args))
    print(kwargs)
    
# func(1, 2, 3, 4, 5, 6, 7, 'a', 'b', 'c')
a = [1,2,3,4,5] # iterable
b = dict(a=1, b=2, c=3)



# func(*a)
# func(*'ê°€ë‚˜ë‹¤ë¼ë§ˆë°”ì‚¬')
# func(*'DIVX')

# func(*a, a=1, b=2, c=3)
func(*a, **b)
```

```python
# ì›¹ìº ì—ì„œ ì½ì€ ë™ì˜ìƒì„ íŒŒì¼ë¡œ ì €ìž¥
import cv2

cap = cv2.VideoCapture(0)

# ìž…ë ¥ ì˜ìƒì˜ size
_, img = cap.read()

# print(img.shape)
size = (img.shape[1], img.shape[0]) # width, height

# print(size)
codec = cv2.VideoWriter_fourcc(*'MJPG')
FPS = cap.get(cv2.CAP_PROP_FPS)
filepath = 'webcam_output.avi'

# VideoWriter ê°ì²´ ìƒì„±
writer = cv2.VideoWriter(filepath, codec, FPS, size)

while True:
    success, image = cap.read()
    if not success:
        print('ìº¡ì³ì‹¤íŒ¨')
        break
    
    cv2.imshow('frame', image)
    writer.write(image)
    if cv2.waitKey(1) == 27:
        break
        
cap.release()
writer.release()
cv2.destroyAllWindows()
cv2.waitKey(1)



```
