# <span style='background-color:#fff5b1'>구현

```python
import tensorflow as tf
import tensorflow_hub as hub

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## <span style='background-color:#fff5b1'>모델 로딩

```python
# 모델을 Tensorflow hub로 부터 다운로드해서 메모리로 로딩.
model = hub.load("https://tfhub.dev/google/movenet/singlepose/thunder/4")
print(type(model))
# 여러 모델들을 담고 있는 객체
```

```python
model.signatures # 여러모델들이 key - 모델 형태로 저장됨. 이것을 확인 (이름 - 모델: signature)
```

```python
movenet = model.signatures['serving_default'] # 모델을 조회
print(type(movenet))
```

## <span style='background-color:#fff5b1'>추론할 이미지 로딩및 전처리
- 이미지를 모델에 맞게 resize한다.
    - thunder: 256, 256
    - lightning: 192, 192
- 타입: int32

```python
# 추론할 이미지 로딩 + 전처리 (resize: 256 X 256, int32 타입.)
image_path = 'test_image/babel1.jpg'

image_orig = cv2.imread(image_path)
print(image_orig.shape)
cv2.imshow('frame', image_orig)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# BGR => RGB + bath axis 추가
input_tensor = cv2.cvtColor(image_orig, cv2.COLOR_BGR2RGB)[np.newaxis, ...]
# resize : thunder - 256 X 256
input_tensor = tf.image.resize_with_pad(input_tensor, 256, 256) # 원본 이미지의 종횡비를 유지하면서 resize
# int32 로 형변환
input_tensor = tf.cast(input_tensor, dtype = tf.int32)

print('원본 크기, dtype:', image_orig.shape, image_orig.dtype)
print('input tensor 크기, dtype:', input_tensor.shape, input_tensor.dtype)
```

```python
print(type(input_tensor))
plt.imshow(input_tensor.numpy()[0].astype('uint8'));
```

## 추론 및 결과 조회

```python
outputs = movenet(input_tensor)
print(type(outputs))
```

```python
outputs.keys()
```

```python
keypoints = outputs['output_0']
print(type(keypoints), keypoints.shape)
# (1:이미지 개수, 1: 사람수, 17:keypoint개수, 3: yxc)
```

```python
nose_kp = keypoints[0, 0, 0] # 코의 keypoint
nose_kp
```

```python
nose_y = nose_kp[0] * image_orig.shape[0]
nose_x = nose_kp[1] * image_orig.shape[1]
nose_conf = nose_kp[2]
print(nose_y.numpy(), nose_x.numpy(), nose_conf.numpy())
```

## 추론한 keypoint 원본 영상에 표시
- keypoints: 찾은 관절표시
- edge: 관절 연결

```python
# 연결할 keypoint 들을 튜플로 묶은 것.
# (0,1) - 0번과 1번 keypoint 들을 연결.
EDGE_LINK = [
    (0, 1),
    (0, 2),
    (1, 3),
    (2, 4),
    (0, 5),
    (0, 6),
    (5, 7),
    (7, 9),
    (6, 8),
    (8, 10),
    (5, 6),
    (5, 11),
    (6, 12),
    (11, 12),
    (11, 13),
    (13, 15),
    (12, 14),
    (14, 16)
]
```

```python
h, w, _ = image_orig.shape
h = w = max(h, w) # 원본 이미지의 height과 width를 큰 쪽으로 통일. => 추론할때 width와 height가 같은 크기로 추론했기 때문에 맞춰준다.
```

```python
frame = cv2.cvtColor(image_orig, cv2.COLOR_BGR2RGB)
frame = tf.image.resize_with_pad(frame, h, w).numpy().astype('uint8') # (256, 256) ==> (562, 562)
plt.imshow(frame)
```

```python
#keypoint 들을 표사
score_threshold = 0.3 # 지정한 confidence socre 이상인 keypoint 들만 출력
denormalized_keypoints = keypoints.numpy().squeeze() * np.array([h, w, 1]) # size 가 1인 값을 소거 (1, 1, 17, 3) ==> (17, 3)

for kp_y, kp_x, kp_score in denormalized_keypoints: # keypoint 좌표에 점(원)을 찍기
    if kp_score >= score_threshold:
        cv2.circle(frame, center = (int(kp_x), int(kp_y)), radius = 3, color = [255,255,0], thickness = -1,
                  lineType = cv2.LINE_AA)
```

```python
plt.figure(figsize = (10, 10))
plt.imshow(frame)
```

```python
# Edge를 출력 => Edge: keypoint간의 연결선 # EDGE_LINK 리스트를 이용
for edge in EDGE_LINK:
    p1, p2 = edge
    y1, x1, score1 = denormalized_keypoints[p1]
    y2, x2, score2 = denormalized_keypoints[p2]
    y1, x1, y2, x2 = int(y1), int(x1), int(y2), int(x2)# 좌표를 정수로 변환
    
    # 두 keypoint의 confiendce score가 threshold 이상이면 그린다.
    if score1 >= score_threshold and score2 >= score_threshold:
        cv2.line(frame, pt1 = (x1, y1), pt2 = (x2, y2), color = [0, 255, 255], thickness= 1, lineType = cv2.LINE_AA)
```

```python
plt.figure(figsize = (7, 7))
plt.imshow(frame)
plt.show()
```
